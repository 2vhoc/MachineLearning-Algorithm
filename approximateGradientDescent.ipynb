```python
import numpy as np
#vd: fx = x**2 + 5*np.sin(x) 
```

```python
def f(x):
    return x**2 + 5*np.sin(x) 
def commaY(x):
    eps = 1e-3
    return (f(x + eps) - f(x - eps))/(2 * eps)
def gradient(X0, learningRate, epochs):
    x = [X0]
    for it in range(epochs):
        xNew = x[-1] - learningRate * commaY(x[-1])
        x.append(xNew)
        if abs(commaY(xNew)) < 1e-3:
            break
    return x, it

##
x, it = gradient(-5, 0.05, 100)
print(x, it)
```

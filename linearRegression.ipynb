```python
df = pd.read_csv('weight-height.csv')
df.columns = ['Gender', 'H', 'W']
```

```python
x = df['H'].values
y = df['W'].values
N = x.shape[0]
```

```python
m = (N*np.sum(x*y) - np.sum(x)*np.sum(y))/(N*np.sum(x**2) - np.sum(x)**2)
b = (np.sum(y) - m*np.sum(x))/N
m, b
```

```python
#recipe:
key = 70
predict = key*m + b
predict
```

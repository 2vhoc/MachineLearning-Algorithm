```python
def commaY(x):
    return x - 1
def y(x):
    return 0.5 * (x-1)**2 - 2
def gradientDescent(X0, learningRate, epochs):
    x = [X0]
    for it in range(epochs):
        xNew = x[-1] - learningRate * commaY(x[-1])
        x.append(xNew)
        if abs(commaY(xNew)) < 1e-3:
            break
        
    return x, it


####
X0 = 5
learningRate = 0.07
epochs = 100

x, it = gradientDescent(X0, learningRate, epochs)
print(x, it)
        
        
```

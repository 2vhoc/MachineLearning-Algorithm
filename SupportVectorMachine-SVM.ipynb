```python
data = pd.read_csv("/kaggle/input/titanic-dataset/Titanic-Dataset.csv")
```

```python
data.head(5)
```

```python
data = data.drop(['PassengerId', 'Name', 'Cabin', 'Embarked', 'Ticket'], axis=1)
```

```python
data
```

```python
data['Sex'] = data['Sex'].map({'male': 1, 'female':0})
```

```python
data.head()
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
#Pclass	Sex	Age	SibSp	Parch	Fare
```

```python
import numpy as np
```

```python
X = data.iloc[:, 1:7]
X = np.array(X)
y = np.array(data.iloc[:, 0])
y
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2)
```

```python
Xtrain.shape
print(Xtrain[:5])
```

```python
# Chênh lệch giữa tuổi và giá trị Sibsp, pclass nó quá chênh lệch nhau nên phải dùng scaler
from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler()
Xtrain = Scaler.fit_transform(Xtrain)
Xtest = Scaler.transform(Xtest)
Xtrain
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import classification_report # chỉ để hiển thị kết quả
SVM = SVC()
paramGrid = {
    'C' : [0.1, 1, 10],
     'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'degree': [2, 3, 4]

}
```

```python
gridSearch = GridSearchCV(estimator=SVM,cv=10, param_grid=paramGrid, )
gridSearch.fit(Xtrain, ytrain)
```

```python
bestPagram = gridSearch.best_params_
bestScore = gridSearch.best_score_
bestPagram, bestScore
```

```python
Model = SVC(**bestPagram)
Model.fit(Xtrain, ytrain)
```

```python
ypre = Model.predict(Xtest)
```

```python
report = classification_report(ypre, ytest)
print(report)
```

```python
xReality = np.array([1, 1, 19, 2, 1, 71.000])

xReality = Scaler.transform(xReality.reshape(1, -1))
print(xReality)
```
